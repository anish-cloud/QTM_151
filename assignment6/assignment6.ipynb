- Count how many cars have horsewpower $> 100$.

````python
# Write your answer here

features = pd.read_csv("/Users/anishnayak/Documents/GitHub/QTM_151/assignment6/Data/features.csv")


features['horsepower'] = pd.to_numeric(features['horsepower'], errors='coerce')

check_horsepower = lambda x: x > 100

horsepower_check_list = features['horsepower'].apply(check_horsepower)

count_horsepower_above_100 = horsepower_check_list.sum()

print(f"Cars with horsepower > 100: {count_horsepower_above_100}")


````
